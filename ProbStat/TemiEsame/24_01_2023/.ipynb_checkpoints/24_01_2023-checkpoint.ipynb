```python
#Import the useful packages
import pandas as pd 
import numpy as np 
import scipy
import seaborn as sns 
import matplotlib.pyplot as plt
from fitter import Fitter
```

**Phase 1: preliminary analysis on the dataset**

```python
#Read the dataset 
df = pd.read_csv("GERD_SCIENCE_17012023090805122.csv")
```

```python
df.head(10)
```

```python
df_2000 = df[df.Year < 2000]
```

```python
df_2015 = df[df.Year > 2000]
```

```python
#Histplot for the accomodation prices (full)
g = sns.histplot(data=df_2000, x="Value", bins=200, 
                 stat="density", kde=True,
                 cumulative=False, common_norm=False, element="step") 
g.set_xlim([-100, 0.2e6])
g.set_ylim([0, 1e-5])
g.set_title("Expenses before 2000")
```

```python
#Histplot for the values (full)
df_2015.Value.hist(bins=50, density=True)
```

```python
#Boxplot distribution
ax = sns.boxplot(data=df_2015, x="PowerCode", y="Value", fliersize=0)
ax.set_ylim([-1e3, 1e4])
```
